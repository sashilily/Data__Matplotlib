```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
#plotting a simple function

price = [48000, 54000, 57000, 49000, 47000, 45000]
year = [2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
plt.grid()
plt.show()
```

```python
batsman = pd.DataFrame({'index':[2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017],
                       'RG Sharma': [404,362, 404, 372, 433,538,390,482,489,333],
                        'V Kohll': [165, 246, 307, 557, 364, 639, 359, 505, 973, 308]
                       })
```

```python
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohll'])
```

```python
# mltiple plots
plt.plot(batsman['index'], batsman['V Kohll'])
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.show()
```

```python
plt.plot(batsman['index'], batsman['V Kohll'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohll CAREER difference')
plt.xlabel('season')
plt.ylabel('Runs scored')
plt.show()
```

```python
#Adding color and dashed and dotted lines

plt.plot(batsman['index'], batsman['V Kohll'], color='#ADAA19', linestyle='dashdot')#dashdot linestyle
plt.plot(batsman['index'], batsman['RG Sharma'], color='#FC00D6', linestyle='dotted')#dotted lines

plt.title('Rohit Sharma Vs Virat Kohll CAREER difference')
plt.xlabel('season')
plt.ylabel('Runs scored')
plt.show()
```

```python
#Line width and marker(size)

plt.plot(batsman['index'], batsman['V Kohll'], color='#ADAA19', 
         linestyle='solid', linewidth=3, 
        marker='>', markersize=10)
plt.plot(batsman['index'], batsman['RG Sharma'], color='#FC00D6', 
         linestyle='dashdot', linewidth=2,
        marker='D')#dotted lines

plt.title('Rohit Sharma Vs Virat Kohll CAREER difference')
plt.xlabel('season')
plt.ylabel('Runs scored')
plt.show()
```

```python
#Legend --> Location

plt.plot(batsman['index'], batsman['V Kohll'], color='#ADAA19', 
         linestyle='solid', linewidth=3, 
        marker='>', markersize=10, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#FC00D6', 
         linestyle='dashdot', linewidth=2,
        marker='D', label='Sharma')

plt.title('Rohit Sharma Vs Virat Kohll CAREER difference')
plt.xlabel('season')
plt.ylabel('Runs scored')

#plt.legend() # Loc= best --> default
plt.legend(loc='upper right')#chosen position
plt.show()
```

### Bar Chart

```python
children = [10, 20, 40, 10, 30]
colors = ['red','blue', 'green', 'yellow', 'pink']

plt.bar(colors, children, color='pink')
```

```python
#horizontal bar chart

plt.barh(colors, children, color='gold')
plt.show()
```

```python
batsman = pd.DataFrame({
    'batsman': ['Hello', 'Morn', 'Even','late','born','karma', 'hell'],
    'index':[2008, 2009, 2010, 2011, 2012, 2013, 2014],
    'RG Sharma': [404,362, 404, 372, 433,538,390],
    'V Kohll': [165, 246, 307, 557, 364, 639, 359]
    })
```

```python
batsman
```

```python
plt.bar(batsman['batsman'], batsman['V Kohll'], color='#DAF7A6')
plt.show()
```

```python
# Multiple bar plots

plt.bar(np.arange(batsman.shape[0]) - 0.2, batsman['V Kohll'], width=0.2)
#plt.bar(np.arange(batsman.shape[0]), batsman['index'], width=0.2)
plt.bar(np.arange(batsman.shape[0]), batsman['RG Sharma'], width=0.2)

plt.show()
```

```python
batsman.shape[0]
```

```python
plt.bar(colors,children, color='black')
plt.xticks(rotation='vertical') #use vertical

plt.show()
```

```python
#Stacked bar:

plt.bar(batsman['batsman'],  batsman['V Kohll'], label = 'V Kohll')
#plt.bar(np.arange(batsman.shape[0]), batsman['index'], width=0.2)

plt.bar(batsman['batsman'], batsman['RG Sharma'], bottom=batsman['V Kohll'], label='RG Sharma')

plt.legend()
plt.show()
```

```python
# On data

df= pd.DataFrame({
    'batsman':['Hello', 'Morn', 'Even','late','born','karma', 'hell'],
    'runs':[31, 175, 70, 2, 33, 11, 100]
})
df
```

```python
plt.pie(df['runs'], labels=df['batsman'])
plt.show()
```

```python
#Percantages
plt.pie(df['runs'], labels=df['batsman'],
        autopct='%0.1f%%',
       colors=['blue', 'gold', 'green', 'orange', 'cyan','pink', 'purple'])
plt.show()
```

```python
# Explode Shadow

plt.pie(df['runs'], labels=df['batsman'],
        autopct='%0.1f%%',
       colors=['blue', 'gold', 'green', 'orange', 'cyan','pink', 'purple'],
       explode=[0.3, 0.1, 0.2, 0.1, 0.2, 0.1, 0.2],
       shadow=True)
plt.show()
```
